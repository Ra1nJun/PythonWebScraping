```python
import os
import requests
from dotenv import load_dotenv
import json
import pandas as pd
```

```python
load_dotenv()

# 환경 변수에서 값 읽기
client_id = os.getenv("NAVER_CLIENT_ID")  # .env 파일의 NAVER_CLIENT_ID
client_secret = os.getenv("NAVER_CLIENT_SECRET")  # .env 파일의 NAVER_CLIENT_SECRET

headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret,
}

# query string 문자열을 dict 선언  
def search_books(query):   
    # query string 문자열을 dict 선언
    payload = {
    'query':  query, #'파이썬',
    'display': 50,
    'sort': 'sim'
    }
    
    url = 'https://openapi.naver.com/v1/search/book.json' #?query=파이썬&display=100&sort=sim

    res = requests.get(url, params=payload, headers=headers)

    items_data = res.json()['items']

    items_list = list()
    for item in items_data:
        item_dict={}
        item_dict['title'] = item['title']
        item_dict['link'] = item['link']
        item_dict['image'] = item['image']
        item_dict['author'] = item['author']
        item_dict['discount'] = item['discount']
        item_dict['publisher'] = item['publisher']
        item_dict['pubdate'] = item['pubdate']
        item_dict['isbn'] = item['isbn']
        item_dict['description'] = item['description']

        items_list.append(item_dict)

    #1.질문 :  검색어로  찾은  책 목록을 json 파일로 저장하기
    if not os.path.isdir('data'): 
        os.mkdir('data')
    with open('data/books.json','w',encoding="utf-8")as file:
        json.dump(items_list, file)


search_books('파이썬썬')

#2.books.json 파일을 Pandas DataFrame로 저장하기
book_df = pd.read_json('data/books.json')

#3.질문 :  검색어로  찾은  책 목록 출력하기
display(book_df)

#4. 질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
display(book_df.loc[book_df['discount']>=20000,  ['title','author','discount','publisher','pubdate']]\
        .sort_values(by='discount', ascending=False).reset_index(drop=True))

#6. 질문 :  검색어로  찾은  책 목록 중에서 출판사가 인피니티북스인 책만 출력하기
display(book_df.loc[book_df['publisher']=='인피니티북스', book_df.columns.drop(['image','description'])].reset_index(drop=True) )
```
